```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("CCPP.csv")
```

```python
x=df[['AT','V','AP','RH']]
y=df['PE']
```

```python
from sklearn.preprocessing import StandardScaler

# SVR is sensitive to feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2)
```

```python
from sklearn.svm import SVR
kernel_matrix=['rbf','linear','poly']
for i in range(len(kernel_matrix)):
    model =SVR(kernel=kernel_matrix[i])
    model.fit(x_train,y_train)
    print(f"The score for {kernel_matrix[i]} is {model.score(x_test,y_test)}")
```

```python
from sklearn.model_selection import cross_val_score

C_values = [0.1, 1, 10, 100, 1000]
cv_scores = []

for C in C_values:
    svr = SVR(kernel='rbf', C=C, epsilon=0.05, gamma='scale')
    scores = cross_val_score(svr, x_train, y_train, cv=5, scoring='r2')
    cv_scores.append(scores.mean())
    print(f"C={C:6} → Mean CV R²: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error

#Let the kernel is rbf using previous results
model=SVR(kernel='rbf',C=100)
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
r2 =r2_score(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
mae= mean_absolute_error(y_test,y_pred)
```

```python
print("Super vector regressor Performance")
print("-----------------------------------")
print(f"MAE  : {mae}")
print(f"RMSE : {rmse}")
print(f"R²   : {r2}")
```
