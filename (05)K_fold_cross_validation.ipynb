```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv("CCPP.csv")
```

```python
x=df[['AT','V','AP','RH']]
y=df['PE']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.model_selection import KFold
```

```python
#I choose 5 different combinations to train the data set
kf =  KFold(n_splits=5)
kf
```

```python
def get_score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
Linear_list=cross_val_score(LinearRegression(),x,y)
print(f"The score for Linear regression is {np.average(Linear_list)}")
```

```python
Decision_list=cross_val_score(DecisionTreeRegressor(),x,y)
print(f"The score for Linear regression is {np.average(Decision_list)}")
```

```python
Rf_list=cross_val_score(RandomForestRegressor(),x,y)
print(f"The score for random forset regressor is {np.average(Rf_list)}")
```
