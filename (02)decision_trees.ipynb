```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df=pd.read_csv("CCPP.csv")
df
```

```python
x=df[['AT','V','AP','RH']]
y=df['PE']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Apply decision tree regression for the data set
dt_model =DecisionTreeRegressor(max_depth=6, random_state=42)
dt_model.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt

# Range of max_depth values to test
max_depth_values = range(1, 21)  # Test depths 1 to 20

# Store R² scores
r2_scores = []

for depth in max_depth_values:
    # Create Decision Tree with current depth
    dt = DecisionTreeRegressor(max_depth=depth, random_state=42)
    dt.fit(x_train, y_train)
    
    # Predict on test set
    y_pred = dt.predict(x_test)
    
    # Calculate R²
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)

# Plot max_depth vs R²
plt.figure(figsize=(10,6))
plt.plot(max_depth_values, r2_scores, marker='x')
plt.title("Decision Tree Regressor: Max Depth vs R² Score")
plt.xlabel("Max Depth")
plt.ylabel("R² Score")
plt.grid(True)
plt.show()
```

```python
#choosing maximum depth as 9
dt = DecisionTreeRegressor(max_depth=9, random_state=42)
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
    
```

```python
r2 =r2_score(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
mae= mean_absolute_error(y_test,y_pred)
```

```python
print("Decision Tree Regressor Performance")
print("-----------------------------------")
print(f"MAE  : {mae}")
print(f"RMSE : {rmse}")
print(f"R²   : {r2}")
```
